##### Relación Presión Sanguínea 

 

La presión arterial se mide en dos valores: la presión sistólica y la presión diastólica. La presión sistólica es la presión en las arterias cuando el corazón late, mientras que la presión diastólica es la presión en las arterias cuando el corazón está en reposo entre latidos. En mujeres embarazadas, la presión arterial normal es de alrededor de 120/80 mmHg. La hipertensión gestacional se diagnostica cuando la presión arterial sistólica es mayor o igual a 140 mmHg y/o la presión arterial diastólica es mayor o igual a 90 mmHg después de la semana 20 de embarazo. La presión arterial alta en el embarazo puede causar complicaciones como la preeclampsia y el nacimiento prematuro


```python
sns.histplot(df, x="SystolicBP",y="DiastolicBP", hue="HeartRate", element="step", palette="flare",) .set_title("Relación entre la Presión Sistólica, diastólica y el ritmo cardiaco");
```

A continuación, se presenta una gráfica donde podemos apreciar la relación entre la temperatura corporal y la concentración de glucosa en sangre

##### Relación entre la Presión Sistólica, Presión Diastólica y Glucosa en Sangre 

A continuación, se presenta una gráfica donde podemos apreciar la relación entre la Presión Sistólica, Presión Diastólica y Glucosa en Sangre

#### Matplotlib

##### Relación entre la Edad y la Glucosa en Sangre
A continuación, se presenta una gráfica donde podemos apreciar la relación entre la Edad y la Glucosa en Sangre. Notemos que como se mostro en la gráfica de edad, tenemos una mayor concentración de población en el rango entre 18-25, por lo cual podemos observar un mayor numero de concentración.

##### Ritmo Cardiaco 

A continuación, se presenta una gráfica donde podemos apreciar el Ritmo cardiaco de las mujeres embarazadas. Podemos observar de forma clara que el ritmo ha aumentando entre 15 - 20 pulsaciones más de lo que se espera para una persona promedio. 

```python
fig, ax = plt.subplots()
ax.step(x, df["HeartRate"], linewidth=2, c='pink' );
plt.title("Grafica Ritmo Cardiaco")
plt.xlabel("Mujeres Embarazadas")
plt.ylabel("Ritmo Cardiaco");
```

##### Relación entre la Edad y la Glucosa en Sangre

##### Glucosa en sanagre 
A continuación se muestra la relación de la Glucosa en sangre de las mujeres embarazdas que cornformaron esta base de datos. Podemos observar que en la mayoría de las mujeres se presentan valores saludables de Glucosa en sangre, que corresponde a 5,6 mmol/L

#### Conclusiones

El embarazo es una etapa de la vida de la mujer que requiere cuidados especiales, ya que existen diversos factores que pueden generar riesgos tanto para la madre como para el feto. Entre los factores de riesgo más importantes se encuentran la presión arterial alta, el ritmo cardiaco, la concentración de glucosa en sangre, la temperatura y la edad. La presión arterial alta no controlada puede provocar daño en los riñones de la madre y aumentar el riesgo de tener preeclampsia o un bebé con peso bajo. Además, las mujeres con problemas cardíacos congénitos tienen un mayor riesgo de tener un bebé con algún tipo de defecto cardíaco. Los niveles altos de glucosa en sangre pueden ser perjudiciales para el bebé durante las primeras semanas de embarazo, incluso antes de saber que está embarazada. La hipertensión en el embarazo es otra complicación que puede tener graves consecuencias tanto para la madre como para el feto. Por lo tanto, es importante que las mujeres embarazadas se realicen controles periódicos y sigan las recomendaciones de su médico para prevenir y controlar estos riesgos.
